Welcome to Day 5! Today, we‚Äôll dive into the **core data structures in Spark: RDDs (Resilient Distributed Datasets), DataFrames, and Datasets.** These structures form the foundation for processing data in distributed environments.

---



### **1Ô∏è‚É£ Resilient Distributed Dataset (RDD) Definition**: RDD is the fundamental data structure in Spark, representing an immutable, fault-tolerant collection of elements that can be processed in parallel.

**Key Features:**

+ **Immutable**: Once created, RDDs cannot be modified. New RDDs can only be generated through transformations.

+ **Fault-Tolerant**: Spark can automatically reconstruct RDDs in case of failure using lineage information.

+ **Lazy Evaluation**: Transformations are not executed immediately but only when an action is triggered.

+ **Low-Level Operations**: Provides fine-grained control over data transformations.

üîé When to use RDDs:

1. ‚úÖ When complete control over data transformations is required.

2. ‚úÖ Suitable for unstructured or semi-structured data.

---


### **2Ô∏è‚É£ DataFrame Definition**: DataFrame is a distributed collection of data organized into named columns, similar to a table in a relational database.

**Key Features:**

+ **Schema-Based**: Designed for structured data with a clear schema, making it easy to work with row-and-column formats.

+ **Optimized**: Uses the Catalyst Optimizer to enhance execution performance.

+ **High-Level Operations**: Supports SQL-like queries and high-level data transformations.

üîé When to use DataFrames:

1. ‚úÖ When working with structured data (row-and-column format).

2. ‚úÖ For high-level operations or SQL-like syntax.

----


### **3Ô∏è‚É£ Dataset**: A Dataset is a strongly-typed collection of data that combines the type-safety of RDDs with the optimizations of DataFrames.

**Key Features:**

+ **Type-Safe**: Enables compile-time type-checking.

+ **Performance Optimization**: Leverages Catalyst Optimizer and Tungsten Execution Engine for high performance.

+ **Supports Object-Oriented Programming**: Ideal for complex data processing pipelines.

üîé When to use Datasets:
1. ‚úÖ Suitable for structured data processing requiring type safety.
2. ‚úÖ Primarily used with Scala and Java (limited support in Python).


###üîë Summary:

####+ RDDs: Flexible and low-level, best for unstructured data.

####+ DataFrames: High performance with high-level operations, ideal for structured data.

#### + Datasets: Combines type safety with performance optimization, best suited for Scala/Java.

## 1. Create an RDD and perform transformations

```python
rdd = spark.sparkContext.parallelize([1, 2, 3, 4])
transformed_rdd = rdd.map(lambda x: x*2)
print(transformed_rdd.collect())
```
